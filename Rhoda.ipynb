```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("bmi_train.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.plot(kind='scatter', x="Height", y="Weight", s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
data.isnull().sum()
```

```python
x=data[["Height"]]
y=data[["Weight"]]
```

```python
model=linear_model.LinearRegression()
```

```python
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test, y_test, label='Actual Values', color='green', marker='o')

# Plot predicted values
plt.plot(x_test, y_pred, label='Predicted Values', color='maroon', marker='x')
```

```python
model.intercept_
```

```python
model.coef_
```

```python
predicted_weigtht= model.predict([[5]])
print(predicted_weigtht)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("Mean squared error")
print(mean_squared_error(y_test,y_pred))
print("R_squared")
print(r2_score(y_test,y_pred))
print("Mean absolute error")
print(mean_absolute_error(y_test,y_pred))
```

```python
From the metrics result, we can see our model is okay but with a fit as the r_squared is low at -0.0002.
This means that the our predictor variable weight only explains 0.02% of the variation in the in the childs weight.
The other 99% variation is explained by other factors not included in our data. This means that height alone cannot predict the weight of a person
and might need additional variables to improve the performance.

The mean absolute error of 29.39 shows that on average the predictions made by your linear regression model differ from the actual
values by approximately 29.39. This means that if the actual weight of a person is of a child is 9 the model will predict it to be 10.43 or 7.57

The mean squared error of 1138.39 shows that our algorithm squared error between the actual value and the predicted value is 3.42.
Because the mean squared error squares the errors, it is more sensitive to the outliers, therofore higher values indicate
the presence of outliers in our data. In his case our value is low meaning we did not have outliers in our model.
```
